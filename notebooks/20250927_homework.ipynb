# 教育程度（a8）、個人收入（b1）兩個變數的相關分析

```python
df_1 = df[["a8", "b1"]].copy().rename(columns = { 'a8': 'education', 'b1': 'income' })
```

```python
print(df_1['education'].value_counts())
print('\n')
print(df_1['income'].value_counts())
```

## Filter out "無" from education

```python
df_1_filtered = df_1.query("education not in ['無', '初職', '自修']")
print(df_1_filtered['education'].value_counts())
print('\n')
print(df_1_filtered['income'].value_counts())
```

## Encode for education
Rules：
1. 小學
2. 國中
3. 高中職
4. 大學與專科
5. 研究所以上

```python
import numpy as np

def categorize_education(value):
    category_mapping = {
        '小學肄業': 1,
        '國(初)中': 2,
        '高職': 3,
        '高中職業科': 3,
        '高中普通科': 3,
        '大學': 4,
        '技術學院、科大': 4,
        '二專': 4,
        '五專': 4,
        '三專': 4,
        '空中大學': 4,
        '軍警專科班': 4,
        '空中行專': 4,
        '士官學校': 4,
        '軍警官學校': 4,
        '軍警專修班': 4,
        '博士': 5,
        '碩士': 5
    }
    
    return category_mapping.get(value, 0)

df_1_encoded = df_1_filtered.copy()
df_1_encoded.loc[:, 'education_encoded'] = df_1_encoded['education'].apply(categorize_education)
```

## Encode for education
Rules：
1. 無收入
2. 低收入（未滿2萬元）
3. 中低收入（2-4萬元）
4. 中等收入（4-6萬元）
5. 中高收入（6-8萬元）
6. 高收入（8萬元以上）

```python
import numpy as np

def categorize_income(value):
    income_mapping = {
        '無收入': 1,
        '未滿1萬元': 2,
        '1萬元以上，未滿2萬元': 2,
        '2萬元以上，未滿3萬元': 3,
        '3萬元以上，未滿4萬元': 3,
        '4萬元以上，未滿5萬元': 4,
        '5萬元以上，未滿6萬元': 4,
        '6萬元以上，未滿7萬元': 5,
        '7萬元以上，未滿8萬元': 5,
        '8萬元以上，未滿9萬元': 6,
        '9萬元以上，未滿10萬元': 6,
        '10萬元以上，未滿11萬元': 6,
        '11萬元以上，未滿12萬元': 6,
        '12萬元以上，未滿13萬元': 6,
        '14萬元以上': 6
    }
    
    return income_mapping.get(value, 0)


df_1_encoded.loc[:, 'income_encoded'] = df_1_encoded['income'].apply(categorize_income)
```

## 「完全排除遺漏值」

```python
df_1_drop_missing_value = df_1_encoded.dropna()
```

### Perform Pearson Correlation

```python
from scipy.stats import pearsonr
import numpy as np

from statistic.pearson import pearson_correlation_table

results, corr_coef, p_value = pearson_correlation_table(df_1_drop_missing_value, 'education_encoded', 'income_encoded')
results
```

## 以「數列平均數」置換個人收入的遺漏值，並以此置換後的新變數做與教育程度的相關分析

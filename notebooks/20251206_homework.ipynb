## Q1: 以收入（inc）作為依變項，性別（sex）為自變項，檢驗教育年數（eduy）是否可以當做上述二者的中介變項。
如果有中介關係，計算 (bc - b‘c) / bc ，顯示中介關係占原關係的百分比

```python
df_1 = df[['inc', 'sex', 'eduy']].copy()
```

檢查遺漏值狀況

```python
from statistic.missing_value import check_missing_value

check_missing_value(df_1)
```

發現有遺漏值，處理遺漏值

```python
df_1_cleared = df_1.dropna()
check_missing_value(df_1_cleared)
```

```python
from statistic.linear_regression import mediation_analysis
reports = mediation_analysis(X = df_1_cleared['sex'], Y = df_1_cleared['inc'], Z = df_1_cleared['eduy'])
```

檢視 $Y = b_cX + a_1 $ 迴歸結果

```python
reports['total_effect_summary']
```

小結：$b_c$ 為13300， P value 為0；由於P value < 0.05 ，具統計顯著性

檢視 $Z = b_aX + a_2 $ 迴歸結果

```python
reports['mediator_summary']
```

小結：$b_a$ 為 0.2318， P value 為0.364；由於P value > 0.05 ，不具統計顯著性

### Answer
整體模型的R Square 為0.048，具統計顯著性。

由於 $b_a$ 不具統計顯著性，中介檢驗不通過，停止檢驗。教育年份對收入與性別無中介作用。

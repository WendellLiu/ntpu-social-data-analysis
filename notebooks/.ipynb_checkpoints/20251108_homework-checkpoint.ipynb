## Q1: 使用迴歸模型，以教育年數（eduy）預測收入（inc）

```python
df_1 = df[['eduy', 'inc']].copy()
```

檢驗遺漏值狀態

```python
from statistic.missing_value import check_missing_value

check_missing_value(df_1)
```

發現有遺漏值，刪除資料

```python
df_1_cleared = df_1.dropna()
df_1_cleared
```

進行簡單線性迴歸分析

```python
from statistic.linear_regression import simple_linear_regression_report

report_1 = simple_linear_regression_report(y_series = df_1_cleared['inc'], x_series = df_1_cleared['eduy'])

print(report_1)
```

### Answer

首先R Square = 0.065 表明此線性模型解釋力弱

接下來對模型是否有預測能力做檢定：  
$H_0: \beta = 0$  
$H_1: \beta \neq 0$  
$\alpha = 0.05$

報告中針對eduy 係數(b)為2059.84 的T 檢定P Value 為0.000。由於P Value < 0.05 ，拒絕虛無假設，此係數不為零，模型有預測能力。

結論：  
1. 此線性模型解釋能力弱
2. 模型係數不為0
3. 依據模型，控制變數後，教育年份每上升一單位，收入上升2059.84 單位

## Q2: 使用迴歸模型，以社交性（sociablity）預測收入

```python
df_2 = df[['sociability', 'inc']].copy()
```

檢驗遺漏值狀態

```python
from statistic.missing_value import check_missing_value

check_missing_value(df_2)
```

發現有遺漏值，刪除資料

```python
df_2_cleared = df_2.dropna()
df_2_cleared
```

進行簡單線性迴歸分析

```python
from statistic.linear_regression import simple_linear_regression_report

report_2 = simple_linear_regression_report(y_series = df_2_cleared['inc'], x_series = df_2_cleared['sociability'])

print(report_2)
```

### Answer

首先R Square = 0.007 表明此線性模型解釋力幾乎為零

接下來對模型是否有預測能力做檢定：  
$H_0: \beta = 0$  
$H_1: \beta \neq 0$  
$\alpha = 0.05$

報告中針對sociability 係數(b)為1107.13 的T 檢定P Value 為0.018。由於P Value < 0.05 ，拒絕虛無假設，此係數不為零，模型有預測能力。

結論：  
1. 此線性模型解釋能力極弱
2. 模型係數不為0
3. 依據模型，控制變數後，社交性分數每上升一單位，收入上升1107.13 單位

## Q3: 控制人際網路（net）之後，社交性（sociability）和收入（inc）之間的淨相關如何？與未控制之前的零階相關差別為何？

```python
df_3 = df[['net', 'sociability', 'inc']].copy()
df_3
```

考察人際網路、社交性、收入等三資料屬性，其中社交性由於alpha（Cronbach’s alpha）0.574為偏低，故將sociability 視為排序變項。  
因此採用spearman 方法做相關分析。

```python
from statistic.correlation import zero_order_correlation_matrix, partial_correlation_matrix
```

```python
zero_order_correlation_matrix(df_3, ['sociability', 'inc'], method='spearman')
```

```python
partial_correlation_matrix(df_3, ['sociability', 'inc'], 'net', method='spearman')
```

### Answer
首先觀察零階相關與淨相關的相關係數：
1. sociability 與inc 的零階相關係數為0.085
2. sociability 與inc 的淨相關係數為0.070

接下來檢定零階相關與淨相關  
$H_0: 不存在相關$  
$H_1: 存在相關$

觀察零階相關與淨相關的雙尾顯著值：
1. 零階相關顯著值為0.013
2. 淨相關顯著值為0.04

兩者都拒絕虛無假設，故兩者都存在相關

然而無論是零階相關或淨相關，係數絕對值皆小於 0.1，故統計上**皆無相關性**。
